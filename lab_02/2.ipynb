### 1: Імпорт бібліотек.

```python
import pandas as pd
import numpy as np
import timeit
```

### 2: Очищення та зчитування даних. 
Тут ми зчитаємо дані. Проаналізувавши цей файл можна зрозуміти, що в ньому роздільником є крапка з комою, а пропущені значення позначені знаком питання. Для зручності я одразу це врахувала.

```python
print("Зчитуємо дані... Це може зайняти кілька секунд.")
df = pd.read_csv('data/household_power_consumption.txt', sep=';', na_values=['?'], low_memory=False)

# Частина з чисткою, видаляємо рядки з пропущеними даними
df = df.dropna()

# Перетворюємо Date та Time у зручний формат datetime для подальшої фільтрації
# Формат дати в файлі: dd/mm/yyyy
df['Datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'], format='%d/%m/%Y %H:%M:%S')

print("Очищення завершено! Розмір датасету:", df.shape)
display(df.head())
```

### 3: Вибірки та профілювання часу (timeit)


3.1: Обрати всі записи, у яких загальна активна споживана потужність перевищує 5 кВт.


```python
def task1_high_power(dataframe):
    start_time = timeit.default_timer()
    

    result = dataframe[dataframe['Global_active_power'] > 5.0]
    
    execution_time = timeit.default_timer() - start_time
    print(f"Завдання 3.1 виконано за {execution_time:.5f} секунд. Знайдено записів: {len(result)}")
    
    return result


res1 = task1_high_power(df)


print("\n--- Записи, де загальна активна потужність > 5 кВт ---")
display(res1)
```

3.2: Обрати всі записи, у яких сила струму лежить в межах 19-20 А, для них виявити ті, у яких пральна машина та холодильних споживають більше, ніж бойлер та кондиціонер.


```python
def task2_current_and_appliances(dataframe):
    start_time = timeit.default_timer()
    result = dataframe[
        (dataframe['Global_intensity'] >= 19.0) & 
        (dataframe['Global_intensity'] <= 20.0) & 
        (dataframe['Sub_metering_2'] > dataframe['Sub_metering_3'])
    ]
    execution_time = timeit.default_timer() - start_time
    print(f"Завдання 2 виконано за {execution_time:.5f} секунд. Знайдено {len(result)} записів.")
    return result
print("--- Результат ---")
res2 = task2_current_and_appliances(df)
display(res2)
```

3.3: Обрати випадковим чином 500000 записів (без повторів елементів вибірки), для них обчислити середні величини усіх 3-х груп споживання електричної енергії


```python
def task3_random_sample_means(dataframe):
    start_time = timeit.default_timer()
    # sample(n=..., replace=False) гарантує відсутність повторів
    sample_df = dataframe.sample(n=500000, replace=False)
    means = sample_df[['Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']].mean()
    execution_time = timeit.default_timer() - start_time
    print(f"Завдання 3 виконано за {execution_time:.5f} секунд.")
    print("Середні значення груп споживання:\n", means)
    return means
    
print("--- Результат ---")
res3 = task3_random_sample_means(df)
display(res1)
```

3.4: Обрати ті записи, які після 18-00 споживають понад 6 кВт за хвилину в середньому, серед відібраних визначити ті, у яких основне споживання електроенергії у вказаний проміжок часу припадає на пральну машину, сушарку, холодильник та освітлення (група 2 є найбільшою), а потім обрати кожен третій результат із першої половини та кожен четвертий результат із другої половини.


```python
def task4_complex_filter(dataframe):
    start_time = timeit.default_timer()
    
    # Спочатку фільтруємо за умовами
    filtered_df = dataframe[
        (dataframe['Datetime'].dt.hour >= 18) & 
        (dataframe['Global_active_power'] > 6.0) & 
        (dataframe['Sub_metering_2'] > dataframe['Sub_metering_1']) & 
        (dataframe['Sub_metering_2'] > dataframe['Sub_metering_3'])
    ]
    
    # Ділимо на дві половини
    half_index = len(filtered_df) // 2
    first_half = filtered_df.iloc[:half_index]
    second_half = filtered_df.iloc[half_index:]
    
    # Обираємо кожен 3 та 4 елемент відповідно
    res_first = first_half.iloc[::3]
    res_second = second_half.iloc[::4]
    
    # Об'єднуємо результати
    final_result = pd.concat([res_first, res_second])
    
    execution_time = timeit.default_timer() - start_time
    print(f"Завдання 4 виконано за {execution_time:.5f} секунд. Знайдено {len(final_result)} записів.")
    return final_result

    
print("--- Результат ---")
res4 = task4_complex_filter(df)
display(res4)
```

### 4: Нормування та стандартизація:

Завдання: Пронормувати та стандартизувати вибраний датасет.

```python
# Обираємо лише числові колонки для нормування та стандартизації
numeric_cols = ['Global_active_power', 'Global_reactive_power', 'Voltage', 
                'Global_intensity', 'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']

print("--- Нормування ---")
# Примітка: нормування перетворює дані так, щоб вони лежали в діапазоні від 0 до 1
df_normalized = (df[numeric_cols] - df[numeric_cols].min()) / (df[numeric_cols].max() - df[numeric_cols].min())
display(df_normalized.head())

print("\n--- Стандартизація ---")
# Примітка: стандартизація перетворює дані так, щоб середнє значення дорівнювало 0, а стандартне відхилення - 1
df_standardized = (df[numeric_cols] - df[numeric_cols].mean()) / df[numeric_cols].std()
display(df_standardized.head())
```

### 5: Кореляція Пірсона та Спірмена

Завдання: Підрахувати коефіцієнт Пірсона та Спірмена для двох integer/real атрибутів.

Для цього завдання я візьму атрибути Global_active_power та Global_intensity, оскільки з курсу фізики відомо, що вони мають бути сильно пов'язані.

```python
# Рахуємо кореляцію між двома числовими атрибутами
col1 = 'Global_active_power'
col2 = 'Global_intensity'

# Коефіцієнт Пірсона (примітка: він оцінює лінійну залежність)
pearson_corr = df[col1].corr(df[col2], method='pearson')

# Коефіцієнт Спірмена (примітка: він оцінює монотонну залежність)
spearman_corr = df[col1].corr(df[col2], method='spearman')

print(f"--- Аналіз кореляції між '{col1}' та '{col2}' ---")
print(f"Коефіцієнт Пірсона:  {pearson_corr:.5f}")
print(f"Коефіцієнт Спірмена: {spearman_corr:.5f}")
```

### 6: One Hot Encoding 

Завдання: Провести One Hot Encoding категоріального атрибута.


```python
print("--- Проведення One Hot Encoding ---")

# Створюємо новий категоріальний атрибут "День тижня" з колонки Datetime
df['Day_of_Week'] = df['Datetime'].dt.day_name()
print("Створено категоріальну колонку 'Day_of_Week'. Початкові значення:")
display(df[['Datetime', 'Day_of_Week']].head())

# Застосовуємо One Hot Encoding (відповідна функція get_dummies в pandas)
# Ця функція створить окремі колонки для кожного дня зі значеннями True або False 
df_encoded = pd.get_dummies(df, columns=['Day_of_Week'], prefix='Day')

# Виводимо результат, щоб показати нові створені колонки
encoded_cols = [col for col in df_encoded.columns if col.startswith('Day_')]

print("\nРезультат One Hot Encoding (нові колонки):")
display(df_encoded[encoded_cols].head())
```
